### **Covariance** ###
Covariance is a measure of the joint variability of two random random variables.

*   If the covariance of two random variables is **positive**, then Y will follow the same tend as X.
*   If the covariance of two random variables is **zero or near zero**, then Y does not depend on the trend of X.
*   If the covariance of two random variables is **negative**, Y will follow the opposite trend as X.

```python
# Calucalating Covereince using Python
def cov(X, Y):
    n = len(X)
    mean_x = sum(X) / n
    mean_y = sum(Y) / n
    return sum([(x - mean_x) * (y - mean_y) for x, y in zip(X, Y)]) / (n-1)
```

### **Correlation** ###
Correlation is the measure of linear correlation between two variables.
*   The **signedness** of the correlation of two variables describes how two variables depend or do not depend on each other.
*   The **magnitude** of a correlation between two variables indicates the strength of the trend.

```python
# Calucalating Correlation using Python
def corr(X, Y):
    n = len(X)
    mean_x = sum(X) / n
    mean_y = sum(Y) / n
    covxy = sum([(x - mean_x) * (y - mean_y) for x, y in zip(X, Y)]) / n
    return covxy / (np.std(X) * np.std(Y))
```

## **Random Samples** ##
A simple random sample is a randomly selected subset of a popluation and often used in practice. We can draw conclusions about the popluation mean **μ** using the sample mean $\bar{X} = \frac{1}{n}(X_1+...+X_n)$.

### **Central Limit Theorem (CLT)** ###
CLT indicates that the properly normalized sum of independent random variables tends toward a normal distribution even if the original variables themselves are not normally distributed.

Sample mean: $μ_\bar{x} = \mu$

Sample standard deviation: $σ_\bar{x}^2 = \frac{σ^2}{n}$ (The larger the value of n, the better the approximation)

```python
# Central Limit Theorm Using Python
def clt(Y):
    n = len(Y)
    uy = np.mean(Y)
    sy = np.std(Y)
    return uy, sy/np.sqrt(n)
```
