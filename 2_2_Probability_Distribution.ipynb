# **Probability Distribution**
A probability distribution is the mathematical function that gives the probabilities of occurance of different possible for an experiment.

## **Probability Axioms**
(1) $P(A)≥0 $

(2) $P(S) = 1$

(3) If $(A\cap B) = 0$, then $P(A∪B) = P(A) + P(B)$

## **Conditional Probability**
Likelihood of an event or outcome occuring, based on the occurrence of a pervious event or outcome. Expressed as a ratio of unconditional probabilites.

For any two events $A$ and $B$ with $P(B)>0$, the conditional probability of $A$ given that $B$ has occured is defined by

$P(A|B) = \frac{P(A \cap B)}{P(B)}$

```python
# Conditional Probability Example Using Python
# Define the total number of cards in the deck
total_cards = 52

# Define the number of Hearts in the deck
hearts = 13

# Define the number of Kings in the deck
kings = 4

# Calculate the probability of drawing a Heart (P(B))
probability_heart = hearts / total_cards

# Calculate the probability of drawing a King and a Heart (P(A ∩ B))
probability_king_and_heart = kings / total_cards

# Calculate the conditional probability of drawing a King given that you have drawn a Heart (P(A|B))
conditional_probability = probability_king_and_heart / probability_heart

# Print the result
print("Conditional Probability of drawing a King given a Heart:", conditional_probability)
```

## **Discrete Random Variables**
Understood as a measureable function degined on a probability space that maps from the sample space to the real numbers.
Screenshot (96).png

```python
# Discrete Random Variables on Python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

NUM_ROLLS = 1000

values = [1, 2, 3, 4, 5, 6]
sample = np.random.choice(values, NUM_ROLLS)

# Numpy arrays containing counts for each side
side, count = np.unique(sample, return_counts=True)
probs = count / len(sample)

# Plot the results
sns.barplot(x=side, y=probs)
plt.title(
    f"Discrete Probability Distribution for Fair 6-Sided Die ({NUM_ROLLS} rolls)")
plt.ylabel("Probability")
plt.xlabel("Outcome")
plt.show()
```
