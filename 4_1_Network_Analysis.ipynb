# **4.1 Network Analysis**
A **network** is a collecion of objects (nodes) and the relationships between them (edges).

A **cluster** (community) is a group of nodes that are more densely connected to each other rather than the rest of the network. Clusters allow people to gain a spatio-temporal understanding of large datasets from networks.

## **Graph Models**
A **graph** is a mathematical structure that consists of a set of nodes and a set of edges that connect the nodes. It is represented as a pair of sets $G=(V,E)$, where $V$ is the set of nodes and $E$ is the set of edges.

There are several types of graph models.

*   **Directed Graph** - a graph where the edges have a direction
*   **Undirected Graph** - a graph where the edges do not have a direction
*   **Weighted Graph** - a graph where the edges have a weight
*   **Mixed Graph** - graphs that contain both directed and undirected edges along with weighted and unweighted edges.

### **Graph Features**

*   A **Path** is a sequence of nodes that are connected by edges.
*   A **Cycle** is a path that starts and ends at the same node.
* A **Tree** is a graph that has no cycles.
* A **Forest** is a graph that has no cycles and is connected.
*A **Spanning Tree** is a tree that contains all the nodes in the graph.
*A **Spanning Forest** is a forest that contains all the nodes in the graph.
*An **n-hop** neighborhood of a node is the set of nodes that are at most hops away from the node.
*The **degree** of a node is the number of edges that are connected to the node.




### Network Graph Demo


```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

nodes = [1, 2, 3, 4, 5, 6, 7]

edges = [(1, 2), (2, 3), (3, 1),
         (4, 5), (5, 6), (6, 4),
         (1, 4), (2, 5), (3, 6), (7, 1), (7, 2), (7, 3)]
G.add_nodes_from(nodes)
G.add_edges_from(edges)


pos = {1: (0, 0), 2: (1, 0), 3: (0.5, 1.5),
       4: (1.5, 0), 5: (2.5, 0), 6: (2, 1.5),
       7: (1, 0.75)}

nx.draw(G, pos, with_labels=True, node_size=200, node_color='lightblue')
plt.show()
```

## **Laplacian Matrix**
A graph with $n$ nodes can be represented by a $n  n$ adjacency matrix $A$ where $A_{ij}=1$ if there is an edge between nodes $i$ and $j$ , otherwise $A_{ij}=0$.



*   $A$ is called the **Adjacency Matrix**
*   If A is a **directed graph**, the $A_{ij} =1$ if there is an edge from node $i$ to node $j$, otherwise $A_{ij}=0$.
*   If A is an **undirected graph**, then $A_{ij}=A_{ij}$, and is **symmetric**.
*   If A is a **weighted graph**, the $A_{ij}$ is the weight of the edge between nodes $i$ and $j$. Also follows the same rules.

The **Laplacian Matrix** is a matrix that is used to represent the connectivity of a graph. The Laplacian Matrix $L$ of graph $G$ is defined as:





```python
import networkx as nx
import numpy as np

# Calculate the Laplacian matrix
laplacian_matrix = nx.laplacian_matrix(G).toarray()

# Convert the Laplacian matrix to a NumPy array for easy manipulation
laplacian_matrix = np.array(laplacian_matrix)

print("Laplacian Matrix:")
print(laplacian_matrix)
```

```python
# Calculate the incidence matrix
incidence_matrix = nx.incidence_matrix(G, oriented=True).toarray()

# Convert the incidence matrix to a NumPy array for easy manipulation
incidence_matrix = np.array(incidence_matrix)

print("Incidence Matrix:")
print(incidence_matrix)
```
